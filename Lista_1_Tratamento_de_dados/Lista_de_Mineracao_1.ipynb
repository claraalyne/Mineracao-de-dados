```python
#item a) Ler o dataset fakeTelegram.BR_2022.csv
import pandas as pd

df = pd.read_csv('fakeTelegram.BR_2022.csv')
df.head()
```

```python
#item b) ) Identificar e listar as posições (células) contendo valores faltantes.

nulos = df.isnull().stack() # junta tudo em uma serie (vetor)
nulos = nulos[nulos == True]
nulos.index.tolist()
```

```python
#item c) Contar quantas linhas possuem valores faltantes.
linhas_com_faltantes = df.isnull().any(axis=1).sum()
print(f"A quantidade de linhas que tem pelo menos um valor faltante é de: {linhas_com_faltantes}" )
```

```python
#item d) Para cada coluna (feature), contar quantas linhas possuem valores faltantes.
celulas_faltantes = df.isnull().sum()
print(celulas_faltantes)
```

```python
#item e) Identificar e listar as linhas repetidas (duplicadas).
duplicadas = df[df.duplicated()]
print(duplicadas)

qtd_duplicadas = df.duplicated().sum()
print(f"Quantidade de linhas duplicadas: {qtd_duplicadas}")
```

```python
#item f) Identificar e listar as posições (células) contendo valores que não pertencem ao
#domínio (tipo de dados) esperado.
dtypes = {'date_message': str,
 'id_member_anonymous': str,
 'id_group_anonymous': str,
 'media': str,
 'media_type': str,
 'media_url': str,
 'has_media': bool,
 'has_media_url': bool,
 'trava_zap': bool,
 'text_content_anonymous': str,
 'dataset_info_id': int,
 'date_system': str,
 'score_sentiment': float,
 'score_misinformation': float,
 'id_message': int,
 'message_type': str,
 'messenger': str,
 'media_name': str,
 'media_md5': str}

tipo_errado = []

for coluna, tipo_esperado in dtypes.items():
    for index, valor in df[coluna].items():
        if not isinstance(valor, tipo_esperado) and pd.notnull(valor):
            tipo_errado.append((index, coluna))

tipo_errado
```

```python
#item g)  Crie uma coluna chamada “caracteres” contendo a quantidade de caracteres da
#coluna “text”.

df["caracteres"] = df["text_content_anonymous"].astype(str).apply(len)
print(df["caracteres"])
```

```python
#item h) ) Crie uma coluna chamada “words” contendo a quantidade de palavras da coluna
#“text”.
df["words"] = df["text_content_anonymous"].astype(str).apply(lambda x: len(x.split()))
print(df["words"])
```

```python
#item i) Crie uma coluna chamada “viral” contendo o valor 0 se o texto da mensagem (valor
# do campo “text”) não for encontrado em outras linhas e 1, caso contrário.

df["viral"] = df.duplicated(subset="text_content_anonymous", keep=False).astype(int)
print(df["viral"])
```

```python
#item j)  Crie uma coluna chamada “sharings” contendo a quantidade de vezes que o texto
#armazenado no atributo “text” aparece no dataset
df["sharings"] = df["text_content_anonymous"].map(df["text_content_anonymous"].value_counts())
print(df["sharings"])
```

```python
#item k) Crie uma coluna chamada “sentiment” contendo os valores: -1 para textos negativos,
# 0 para textos neutros e 1 para textos positivos.
df["sentiment"] = df["score_sentiment"].apply(lambda x: -1 if x < 0.05 else (1 if x > 0.05 else 0))
print(df["sentiment"])
```

```python
#item l) Eliminar as linhas cujo valor da coluna “text” contenham “trava-zaps”.

df = df[~df['trava_zap']].reset_index(drop=True)
print(f"Total de linhas que sobraram: {len(df)}")
```

```python
#item m)Identificar inconsistências entre os atributos (features).

incons_1 = df[(df['has_media'] == True) & (df['media'].isna() | (df['media'] == ''))]
incons_2 = df[(df['has_media_url'] == True) & (df['media_url'].isna() | (df['media_url'] == ''))]
incons_3 = df[(df['has_media'] == False) & (df['media_type'].notna()) & (df['media_type'] != '')]
incons_4 = df[(df['score_sentiment'] < -1) | (df['score_sentiment'] > 1)]
incons_5 = df[(df['score_misinformation'] < 0) | (df['score_misinformation'] > 1)]
incons_6 = df[(df['has_media'] == False) & (df['text_content_anonymous'].isna() | (df['text_content_anonymous'] == ''))]
incons_7 = df[(df['trava_zap'] == True) & (df['text_content_anonymous'].str.len() < 50)]

print(f"Inconsistência 1 (has_media True mas media vazia): {len(incons_1)}")
print(f"Inconsistência 2 (has_media_url True mas media_url vazia): {len(incons_2)}")
print(f"Inconsistência 3 (media_type existe mas has_media False): {len(incons_3)}")
print(f"Inconsistência 4 (score_sentiment fora do intervalo): {len(incons_4)}")
print(f"Inconsistência 5 (score_misinformation fora do intervalo): {len(incons_5)}")
print(f"Inconsistência 6 (sem mídia e sem texto): {len(incons_6)}")
print(f"Inconsistência 7 (trava_zap com texto pequeno): {len(incons_7)}")
```
